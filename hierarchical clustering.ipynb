```python
import numpy as np
import scipy.cluster.hierarchy as sch
import matplotlib.pyplot as plt

# Labels
labels = ['P1', 'P2', 'P3', 'P4', 'P5']

# Distance matrix (condensed form)
# SciPy requires condensed distance matrix for linkage
distance_matrix = np.array([
    0.10,  # P1-P2
    0.41,  # P1-P3
    0.55,  # P1-P4
    0.35,  # P1-P5
    0.64,  # P2-P3
    0.47,  # P2-P4
    0.98,  # P2-P5
    0.44,  # P3-P4
    0.85,  # P3-P5
    0.76   # P4-P5
])

# Single Linkage Dendrogram
linked_single = sch.linkage(distance_matrix, method='single')

plt.figure(figsize=(8, 4))
sch.dendrogram(linked_single, labels=labels)
plt.title('Hierarchical Clustering - Single Linkage')
plt.xlabel('Points')
plt.ylabel('Distance')
plt.grid(True)
plt.show()

# Complete Linkage Dendrogram
linked_complete = sch.linkage(distance_matrix, method='complete')

plt.figure(figsize=(8, 4))
sch.dendrogram(linked_complete, labels=labels)
plt.title('Hierarchical Clustering - Complete Linkage')
plt.xlabel('Points')
plt.ylabel('Distance')
plt.grid(True)
plt.show()
```
